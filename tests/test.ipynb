# **Getting Started**

```python
import numpy as np

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from plot_frames import *
from spatialmath import *
```

# **Plotting in 2D**

Plotting a rotation in 2D.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()

dims = [-2,2]
ax.set_xlim(dims)
ax.set_ylim(dims)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.set_xticks(np.arange(min(dims), max(dims)+1, 1.0))
ax.set_yticks(np.arange(min(dims), max(dims)+1, 1.0))

R = SO2(45, unit='deg')
print(R)

plot_rot2d(R, frame='R')
```

Plotting a pose in 2D.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot()

dims = [-2,2]
ax.set_xlim(dims)
ax.set_ylim(dims)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.set_xticks(np.arange(min(dims), max(dims)+1, 1.0))
ax.set_yticks(np.arange(min(dims), max(dims)+1, 1.0))

T = SE2([0,0], unit='deg')
print(T)

plot_pose2d(T, frame='T')
```

# **Plotting in 3D**

Plotting a rotation about the Z axes in 3D.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

dims = [-2,2]
ax.set_proj_type('ortho')
ax.set_xlim(dims)
ax.set_ylim(dims)
ax.set_zlim(dims)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
ax.set_xticks(np.arange(min(dims), max(dims)+1, 1.0))
ax.set_yticks(np.arange(min(dims), max(dims)+1, 1.0))
ax.set_zticks(np.arange(min(dims), max(dims)+1, 1.0))

R = SO3().Rz(45, unit='deg')
print(R)

plot_rot3d(R, frame='R')
```

Ploting a pose in 3D.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

dims = [-2,2]
ax.set_proj_type('ortho')
ax.set_xlim(dims)
ax.set_ylim(dims)
ax.set_zlim(dims)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
ax.set_xticks(np.arange(min(dims), max(dims)+1, 1.0))
ax.set_yticks(np.arange(min(dims), max(dims)+1, 1.0))
ax.set_zticks(np.arange(min(dims), max(dims)+1, 1.0))

T = SE3()
print(T)

plot_pose3d(T, frame='T')
```
