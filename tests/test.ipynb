```python
import os
import sys

import numpy as np
import plot_frames

from matplotlib import pyplot as plt
from spatialmath import *
```

```python
fig = plt.figure()
ax = fig.add_subplot()

dims = [0,2]
ax.set_xlim(dims)
ax.set_ylim(dims)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.set_xticks(np.arange(min(dims), max(dims)+1, 0.5))
ax.set_yticks(np.arange(min(dims), max(dims)+1, 0.5))

T = SE2([0.5,0.5])
print(f"T=\n{T}")

plot_frames.plot_text2d((0.5,0.5), text='A')
plot_frames.plot_arrow2d((0,0), (0.5,0.5), width=0.01)
plot_frames.plot_pose2d(T, frame='A', color="blue")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

dims=[0,2]
ax.set_xlim(dims)
ax.set_ylim(dims)
ax.set_zlim(dims)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.grid(True)
ax.set_xticks(np.arange(min(dims), max(dims)+1, 0.5))
ax.set_yticks(np.arange(min(dims), max(dims)+1, 0.5))
ax.set_zticks(np.arange(min(dims), max(dims)+1, 0.5))

T = SE3([0.5,0.5,0.0])
print(f"T=\n{T}")

plot_frames.plot_text3d((0.5,0.5,0.0), text='A')
plot_frames.plot_arrow3d(np.array([0.0,0.0,0.0]), np.array([0.5,0.5,0.0]))
plot_frames.plot_pose3d(T, frame='T', color="blue")
```
