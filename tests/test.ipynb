```python
import os
import sys

import numpy as np
import plot_frames

from matplotlib import pyplot as plt
from spatialmath import *
```

```python
fig = plt.figure()
ax = fig.add_subplot()

T = SE2([0.5,0.5])
print(f"T=\n{T}")

plot_frames.plot_text2d((0.5,0.5), text='A')
plot_frames.plot_arrow2d((0,0), (0.5,0.5), width=0.01)
plot_frames.plot_pose2d(T, frame='A', color="blue")

ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

T = SE3([0.5,0.5,0.0])
print(f"T=\n{T}")

plot_frames.plot_text3d((0.5,0.5,0.0), text='A')
plot_frames.plot_arrow3d(np.array([0.0,0.0,0.0]), np.array([0.5,0.5,0.0]))
plot_frames.plot_pose3d(T, frame='T', color="blue")

ax.set_xlim([0, 2])
ax.set_ylim([0, 2])
ax.set_zlim([0, 2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
